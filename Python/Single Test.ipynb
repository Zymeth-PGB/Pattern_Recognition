```python
import numpy as np
from PIL import Image

from change_dilation import change_dilation

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Segmentation data
pupil_xyr = [200, 200, 51]
iris_xyr = [196, 210, 181]

# Input dilation
dil1 = pupil_xyr[2]/iris_xyr[2]

# Output dilations
dil2 = 0.15
dil3 = 0.55
```

```python
# Open input image
im1 = Image.open('../samples/LVL.png')
```

```python
# Change Dilation
im2 = change_dilation(im1, dil2, pupil_xyr, iris_xyr)
im3 = change_dilation(im1, dil3, pupil_xyr, iris_xyr)
```

```python
# Show results
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(15, 5))
fig.suptitle('Dilation Comparison')

ax1.imshow(im1)
ax1.axis('off')
ax1.set_title('Original image: Rp/Ri = {:0.3f}'.format(dil1))

ax2.imshow(im2)
ax2.axis('off')
ax2.set_title('Reduced dilation: Rp/Ri = {:0.3f}'.format(dil2))

ax3.imshow(im3)
ax3.axis('off')
ax3.set_title('Increased dilation: Rp/Ri = {:0.3f}'.format(dil3))

plt.show()
```
